```python
def lower_space(title):
  new_title = title.lower().replace(' ','_')
  return new_title
```

## Две колонки с типами комнат. Посчитать сколько раз значения отличаются

### Посмотрим на эти **колонки**

```python
df[['reserved_room_type','assigned_room_type']]
```

### Посчитаем количество различий 1 способом

```python
df.query('reserved_room_type != assigned_room_type')[['reserved_room_type','assigned_room_type']]
```

### Посчитаем количество различий 2 способом

```python
len(df.loc[df.reserved_room_type != df.assigned_room_type])
```
